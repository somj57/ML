```python
heart_disease = pd.read_csv("heart-disease1.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins = 10);
```

```python
heart_disease.plot.hist(figsize = (10,30),subplots = True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind = 'scatter',x='age',y='chol',c='target');
```

```python
fig , ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',x='age',y='chol',c='target',ax=ax)
#ax.set_xlim([45,100]);
```

```python
fig, ax =  plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

ax.set(title = "Heart Diseases", xlabel="Age",ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(over_50["chol"].mean(),linestyle = '--');
```

# Subplots

```python
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10))

scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])
ax0.set(title = "Heart Diseases", xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(y=over_50["chol"].mean(),linestyle = '--');

scatter1 = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
ax1.set(title = "Heart Diseases", xlabel="Age",ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),linestyle = '--');

fig.suptitle('Heart Dieases Analysis', fontsize = 16, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid')

fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10))

scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                         cmap='winter')

ax0.set(title = "Heart Diseases", xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(y=over_50["chol"].mean(),linestyle = '--');
ax0.set_xlim([50,80])
ax0.set_ylim([60,600])

scatter1 = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap='winter')

ax1.set(title = "Heart Diseases", xlabel="Age",ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),linestyle = '--');
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

fig.suptitle('Heart Dieases Analysis', fontsize = 16, fontweight="bold");
```
